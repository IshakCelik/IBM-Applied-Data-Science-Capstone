```python
import folium
from folium.features import DivIcon

# üåç Initialize map centered near the US
nasa_coordinate = [29.559684888503615, -95.0830971930759]
site_map = folium.Map(location=nasa_coordinate, zoom_start=5)

# üîÅ Loop over each launch site and add a circle + text label
for index, row in launch_sites_df.iterrows():
    coordinate = [row['Lat'], row['Long']]
    site_name = row['Launch Site']

    # üü† Add circle with popup
    folium.Circle(
        location=coordinate,
        radius=1000,
        color='#000000',
        fill=True
    ).add_child(
        folium.Popup(site_name)
    ).add_to(site_map)

    # üè∑Ô∏è Add text label with DivIcon
    folium.Marker(
        location=coordinate,
        icon=DivIcon(
            icon_size=(20, 20),
            icon_anchor=(0, 0),
            html=f'<div style="font-size: 12px; color:#d35400;"><b>{site_name}</b></div>'
        )
    ).add_to(site_map)

# üó∫Ô∏è Display the map
site_map
```

Answer:
No ‚Äî not all launch sites are close to the Equator.

The Equator is at 0¬∞ latitude.

Launch sites like:

CCAFS LC-40 and KSC LC-39A are in Florida (lat ‚âà 28.5¬∞N) ‚Äî closer to the Equator

VAFB SLC-4E is in California (lat ‚âà 34.6¬∞N) ‚Äî further from the Equator

Conclusion:
Most SpaceX launch sites are in the Northern Hemisphere, but only the Florida-based sites are relatively close to the Equator. Launching near the equator provides more rotational boost for equatorial orbits, which can reduce fuel use.

üåä 2. Are all launch sites in very close proximity to the coast?
Answer:
‚úÖ Yes ‚Äî all launch sites are located near coastlines.

KSC LC-39A and CCAFS LC-40/SLC-40 are right on the Atlantic coast of Florida.

VAFB SLC-4E is on the Pacific coast of California.

Why?

Being near a coast allows rockets to launch over the ocean, minimizing risk to populated areas in case of failure.

It also simplifies recovery operations for boosters landing on drone ships.

# TASK 3: Calculate the distances between a launch site to its proximities


Next, we need to explore and analyze the proximities of launch sites.


Let's first add a `MousePosition` on the map to get coordinate for a mouse over a point on the map. As such, while you are exploring the map, you can easily find the coordinates of any points of interests (such as railway)


```python
# Add Mouse Position to get the coordinate (Lat, Long) for a mouse over on the map
formatter = "function(num) {return L.Util.formatNum(num, 5);};"
mouse_position = MousePosition(
    position='topright',
    separator=' Long: ',
    empty_string='NaN',
    lng_first=False,
    num_digits=20,
    prefix='Lat:',
    lat_formatter=formatter,
    lng_formatter=formatter,
)

site_map.add_child(mouse_position)
site_map
```

Now zoom in to a launch site and explore its proximity to see if you can easily find any railway, highway, coastline, etc. Move your mouse to these points and mark down their coordinates (shown on the top-left) in order to the distance to the launch site.


You can calculate the distance between two points on the map based on their `Lat` and `Long` values using the following method:


```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
import folium
from folium.features import DivIcon
from folium.plugins import MarkerCluster, MousePosition
from math import sin, cos, sqrt, atan2, radians

# ... (your existing code for calculate_distance, site_map, etc.)

# Coordinates for CCAFS LC-40
launch_site_lat = 28.563197
launch_site_lon = -80.576820

# Estimated coordinates for nearby features
city_lat = 28.53
city_lon = -80.65
railway_lat = 28.57
railway_lon = -80.58
highway_lat = 28.56
highway_lon = -80.56

# Calculate distances
city_distance = calculate_distance(launch_site_lat, launch_site_lon, city_lat, city_lon)
railway_distance = calculate_distance(launch_site_lat, launch_site_lon, railway_lat, railway_lon)
highway_distance = calculate_distance(launch_site_lat, launch_site_lon, highway_lat, highway_lon)

# Function to create marker and line
def add_marker_and_line(feature_lat, feature_lon, distance, feature_name, color):
    # Create marker
    marker = folium.Marker(
        location=[feature_lat, feature_lon],
        icon=DivIcon(
            icon_size=(20, 20),
            icon_anchor=(0, 0),
            html='<div style="font-size: 12; color:#d35400;"><b>%s: %s KM</b></div>' % (feature_name, "{:10.2f}".format(distance)),
        )
    ).add_to(site_map)

    # Create line
    folium.PolyLine(
        locations=[[launch_site_lat, launch_site_lon], [feature_lat, feature_lon]],
        color=color,
        weight=2
    ).add_to(site_map)

# Add markers and lines for each feature
add_marker_and_line(city_lat, city_lon, city_distance, "City", "red")
add_marker_and_line(railway_lat, railway_lon, railway_distance, "Railway", "green")
add_marker_and_line(highway_lat, highway_lon, highway_distance, "Highway", "blue")

# Display the map
site_map
```
